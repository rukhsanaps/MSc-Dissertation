# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("energydata_complete.csv")
```

```python
# Convert 'date' to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python

# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Get top 15 features most correlated with Appliances
top_corr_features = corr_matrix['Appliances'].abs().sort_values(ascending=False).head(15).index

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df[top_corr_features].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Heatmap of Top 15 Features Correlated with Appliances")
plt.tight_layout()
plt.show()
```

## Appliances Energy Consumption over Time

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 4))
sns.lineplot(data=df, x='date', y='Appliances')
plt.title("Appliances Energy Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("Appliances (Wh)")
plt.tight_layout()
plt.show()
```

## Heat map showing energy consumption every hour during different days of the week

```python
df['hour'] = df['date'].dt.hour
df['day_of_week'] = df['date'].dt.dayofweek  # 0=Monday, 6=Sunday

# Create pivot table: rows = day of week, columns = hour, values = average appliances usage
pivot_dow = df.pivot_table(values='Appliances', index='day_of_week', columns='hour', aggfunc='mean')

# Rename day numbers to labels
day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
pivot_dow.index = day_labels

# Plot heatmap
plt.figure(figsize=(12, 5))
sns.heatmap(pivot_dow, cmap='YlOrRd', annot=False)
plt.title("Average Appliance Energy Consumption by Hour and Day of Week")
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")
plt.tight_layout()
plt.show()
```

## Average Energy Consumption per hour (Weekday or Weekend)

```python

# Create 'is_weekend' column: 1 for Saturday and Sunday, 0 otherwise
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

# Group by hour and is_weekend to get average Appliance consumption
avg_by_hour = df.groupby(['hour', 'is_weekend'])['Appliances'].mean().reset_index()

# Map numerical labels to string labels for clarity
avg_by_hour['DayType'] = avg_by_hour['is_weekend'].map({0: 'Weekday', 1: 'Weekend'})

# Plot line chart
plt.figure(figsize=(10, 5))
sns.lineplot(data=avg_by_hour, x='hour', y='Appliances', hue='DayType', marker='o')
plt.title("Average Appliance Energy Consumption per Hour (Weekday vs Weekend)")
plt.xlabel("Hour of Day")
plt.ylabel("Average Appliances (Wh)")
plt.xticks(range(0, 24))
plt.legend(title="Day Type")
plt.tight_layout()
plt.show()
```

## Average Energy Consumption by Day of the week

```python


# Map numeric days to labels
day_labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_name'] = df['day_of_week'].map(dict(enumerate(day_labels)))

# Group by day of week and hour to get average appliance consumption
weekly_hourly_avg = df.groupby(['day_name', 'hour'])['Appliances'].mean().reset_index()

# Set consistent day order for plotting
weekly_hourly_avg['day_name'] = pd.Categorical(weekly_hourly_avg['day_name'], categories=day_labels, ordered=True)

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=weekly_hourly_avg, x='hour', y='Appliances', hue='day_name', marker='o')
plt.title("Hourly Appliance Energy Consumption by Day of Week")
plt.xlabel("Hour of Day")
plt.ylabel("Average Appliances (Wh)")
plt.xticks(range(0, 24))
plt.legend(title='Day of Week')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
hourly_avg = df.groupby('hour')[['Appliances', 'lights']].mean().reset_index()

# Plot stacked area chart
plt.figure(figsize=(12, 6))
plt.stackplot(hourly_avg['hour'], hourly_avg['Appliances'], hourly_avg['lights'],
              labels=['Appliances', 'Lights'], alpha=0.8)
plt.title("Average Hourly Energy Consumption: Appliances vs Lights")
plt.xlabel("Hour of Day")
plt.ylabel("Energy Consumption (Wh)")
plt.xticks(range(0, 24))
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
start_date = "2016-02-01"  # change to any date present in your data
end_date = "2016-02-08"
week_data = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Plot
plt.figure(figsize=(14, 5))
plt.plot(week_data['date'], week_data['Appliances'], label="Appliance Consumption")
plt.title(f"Appliance Energy Consumption from {start_date} to {end_date}")
plt.xlabel("Date and Time")
plt.ylabel("Appliances (Wh)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
